# Decision Tree Model Application

```python
import pandas as pd
```

```python
Mutual_Fund_Data= pd.read_csv('Mutual Fund Returns Data.csv')
Mutual_Fund_Data
```

```python
Mutual_Fund_Data.isnull().sum()
```

```python
# Extraction of independent and dependent variable
X=Mutual_Fund_Data.iloc[:,0:-1 ]
Y=Mutual_Fund_Data.iloc[:,-1: ]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.replace('Direct Plan', 0, inplace=True)
Y.replace('Regular', 1, inplace=True)
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
X= pd.get_dummies(X, drop_first= True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.25, random_state= 2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC_entr=DecisionTreeClassifier(criterion='entropy')
DTC_entr.fit(X_train, Y_train)
```

```python
Y_pred=DTC_entr.predict(X_test)
```

```python
# cm
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, Y_pred)
```

```python
# accuracy----Information Gain ASM
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC_gini=DecisionTreeClassifier(criterion='gini')
DTC_gini.fit(X_train, Y_train)
```

```python
# pred
Y_pred=DTC_gini.predict(X_test)
```

```python
# cm
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, Y_pred)
```

```python
# accuracy----Gini Index ASM
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)*100
```
